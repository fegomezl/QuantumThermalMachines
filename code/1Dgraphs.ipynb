```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Temp50 = pd.read_csv('../results/temp_50.csv')
Temp100 = pd.read_csv('../results/temp_100.csv')
Temp200 = pd.read_csv('../results/temp_200.csv')

Gamma50 = pd.read_csv('../results/gamma_50.csv')
Gamma100 = pd.read_csv('../results/gamma_100.csv')
Gamma200 = pd.read_csv('../results/gamma_200.csv')

Energy20 = pd.read_csv('../results/energy_20.csv')
Energy50 = pd.read_csv('../results/energy_50.csv')
Energy100 = pd.read_csv('../results/energy_100.csv')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3)
fig.tight_layout()
ax1.set_ylabel(r'Particle Current $J_p$')
ax1.grid()
ax2.grid()
ax3.grid()
ax1.set_xscale('log')
ax1.set_xlabel('Temperature')
ax1.plot(Temp50['T'], Temp50['Jₚ'])
ax1.plot(Temp100['T'], Temp100['Jₚ'])
ax1.plot(Temp200['T'], Temp200['Jₚ'])

ax2.set_xlabel(r'System-lead coupling')
ax2.plot(Gamma50['Γ'], Gamma50['Jₚ'])
ax2.plot(Gamma100['Γ'], Gamma100['Jₚ'])
ax2.plot(Gamma200['Γ'], Gamma200['Jₚ'])

ax3.set_xlabel(r'single level energy')
ax3.plot(Energy20['ϵ'], Energy20['Jₚ'])
ax3.plot(Energy50['ϵ'], Energy50['Jₚ'])
ax3.plot(Energy100['ϵ'], Energy100['Jₚ'])
plt.show()
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3)
fig.tight_layout()
ax1.set_ylabel(r'Energy Current $J_h$')
ax1.grid()
ax2.grid()
ax3.grid()
ax1.set_xscale('log')
ax1.set_xlabel('Temperature')
ax1.plot(Temp50['T'], Temp50['Jₕ'])
ax1.plot(Temp100['T'], Temp100['Jₕ'])
ax1.plot(Temp200['T'], Temp200['Jₕ'])

ax2.set_xlabel(r'System-lead coupling')
ax2.plot(Gamma50['Γ'], Gamma50['Jₕ'])
ax2.plot(Gamma100['Γ'], Gamma100['Jₕ'])
ax2.plot(Gamma200['Γ'], Gamma200['Jₕ'])

ax3.set_xlabel(r'single level energy')
ax3.plot(Energy20['ϵ'], Energy20['Jₕ'])
ax3.plot(Energy50['ϵ'], Energy50['Jₕ'])
ax3.plot(Energy100['ϵ'], Energy100['Jₕ'])
plt.show()
```
