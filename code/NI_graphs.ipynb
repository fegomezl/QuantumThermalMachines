```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Energy40 = pd.read_csv('../results/energy_40.csv')
Energy60 = pd.read_csv('../results/energy_60.csv')
Energy80 = pd.read_csv('../results/energy_80.csv')
Energy100 = pd.read_csv('../results/energy_100.csv')
Energy120 = pd.read_csv('../results/energy_120.csv')
Energy140 = pd.read_csv('../results/energy_140.csv')
Energy160 = pd.read_csv('../results/energy_160.csv')
Energy180 = pd.read_csv('../results/energy_180.csv')
Energy200 = pd.read_csv('../results/energy_200.csv')

Temp40 = pd.read_csv('../results/temp_40.csv')
Temp60 = pd.read_csv('../results/temp_60.csv')
Temp80 = pd.read_csv('../results/temp_80.csv')
Temp100 = pd.read_csv('../results/temp_100.csv')
Temp120 = pd.read_csv('../results/temp_120.csv')
Temp140 = pd.read_csv('../results/temp_140.csv')
Temp160 = pd.read_csv('../results/temp_160.csv')
Temp180 = pd.read_csv('../results/temp_180.csv')
Temp200 = pd.read_csv('../results/temp_200.csv')

Gamma40 = pd.read_csv('../results/gamma_40.csv')
Gamma60 = pd.read_csv('../results/gamma_60.csv')
Gamma80 = pd.read_csv('../results/gamma_80.csv')
Gamma100 = pd.read_csv('../results/gamma_100.csv')
Gamma120 = pd.read_csv('../results/gamma_120.csv')
Gamma140 = pd.read_csv('../results/gamma_140.csv')
Gamma160 = pd.read_csv('../results/gamma_160.csv')
Gamma180 = pd.read_csv('../results/gamma_180.csv')
Gamma200 = pd.read_csv('../results/gamma_200.csv')
```

```python
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3)

colors = plt.cm.cividis_r(np.linspace(0,1,20))

#------------------------------------------------------------------

ax1.plot(Energy40[''], Energy40['J'], color=colors[3])
ax1.plot(Energy60[''], Energy60['J'], color=colors[5])
ax1.plot(Energy80[''], Energy80['J'], color=colors[7])
ax1.plot(Energy100[''], Energy100['J'], color=colors[9])
ax1.plot(Energy120[''], Energy120['J'], color=colors[11])
ax1.plot(Energy140[''], Energy140['J'], color=colors[13])
ax1.plot(Energy160[''], Energy160['J'], color=colors[15])
ax1.plot(Energy180[''], Energy180['J'], color=colors[17])
ax1.plot(Energy200[''], Energy200['J'], color=colors[19])

ax1.grid()
ax1.set_xlim(-1,1)
ax1.set_ylim(0, 0.012)
ax1.set_ylabel(r'Particle Current J')
ax1.set_xlabel(r'Single level energy ')
               
#------------------------------------------------------------------

ax2.plot(Temp40['T'], Temp40['J'], color=colors[3])
ax2.plot(Temp60['T'], Temp60['J'], color=colors[5])
ax2.plot(Temp80['T'], Temp80['J'], color=colors[7])
ax2.plot(Temp100['T'], Temp100['J'], color=colors[9])
ax2.plot(Temp120['T'], Temp120['J'], color=colors[11])
ax2.plot(Temp140['T'], Temp140['J'], color=colors[13])
ax2.plot(Temp160['T'], Temp160['J'], color=colors[15])
ax2.plot(Temp180['T'], Temp180['J'], color=colors[17])
ax2.plot(Temp200['T'], Temp200['J'], color=colors[19])
               
ax2.grid()
ax2.set_xlim(1e-4,1e1)
ax2.set_ylim(0, 0.012)
ax2.set_xlabel('Temperature T')
ax2.set_xscale('log')

#------------------------------------------------------------------

ax3.plot(Gamma40[''], Gamma40['J'], color=colors[3])
ax3.plot(Gamma60[''], Gamma60['J'], color=colors[5])
ax3.plot(Gamma80[''], Gamma80['J'], color=colors[7])
ax3.plot(Gamma100[''], Gamma100['J'], color=colors[9])
ax3.plot(Gamma120[''], Gamma120['J'], color=colors[11])
ax3.plot(Gamma140[''], Gamma140['J'], color=colors[13])
ax3.plot(Gamma160[''], Gamma160['J'], color=colors[15])
ax3.plot(Gamma180[''], Gamma180['J'], color=colors[17])
ax3.plot(Gamma200[''], Gamma200['J'], color=colors[19])
               
ax3.grid()
ax3.set_xlim(0, 0.5)
ax3.set_ylim(0, 0.018)
ax3.set_xlabel(r'System-lead coupling ')
               
#------------------------------------------------------------------

cbar = fig.colorbar(plt.cm.ScalarMappable(cmap='cividis_r'), ax=ax3, 
                    ticks=[n/10 for n in range(0,11)])
cbar.ax.set_yticklabels([str(20*n) for n in range(0,11)])  
cbar.set_label('L')

fig.tight_layout()
plt.show()
```
